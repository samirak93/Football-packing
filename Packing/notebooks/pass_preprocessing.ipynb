```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.DataFrame(pd.read_csv('passes.csv'))
df.tail()
```

```python
sender_x=[]
sender_y=[]
receiver_x=[]
receiver_y=[]

for index,value in df_filter.iterrows():
    sender_id=value['sender_id'].astype(int)
    receiver_id=value['receiver_id'].astype(int)

    sender_x.append(value.iloc[sender_id+3])
    sender_y.append(value.iloc[sender_id+31])
    
    receiver_x.append(value.iloc[receiver_id+3])
    receiver_y.append(value.iloc[receiver_id+31])
#     sender_x.append(value.lindex+4)
df_filter['sender_x']=sender_x
df_filter['sender_y']=sender_y
df_filter['receiver_x']=receiver_x
df_filter['receiver_y']=receiver_y
df_filter=df_filter.dropna(subset=['sender_x'])
df_filter=df_filter.dropna(subset=['receiver_x'])

df_filter.head()
```

```python
dist_closest_own_player=[]
dist_closest_opp_player=[]

for index,value in df_filter.iterrows():
    sender_own_players=value['dist_sender_own_players']
    sender_opp_players=value['dist_sender_opp_players']
    #use index 1 since after removing nan, sender-sender distance will be 0 and minimum. so 2nd min distance is at 1
    dist_closest_own_player.append(np.partition(sender_own_players[~np.isnan(sender_own_players)],1)[1])
    #for opposition player, index 0 gives lowest distance between closest player
    dist_closest_opp_player.append(np.partition(sender_opp_players[~np.isnan(sender_opp_players)],1)[0])

df_filter['dist_closest_own_player']=dist_closest_own_player
df_filter['dist_closest_opp_player']=dist_closest_opp_player

df_filter.head()
```

```python
from numpy import linalg as LA
dist_defender_lop=[] 
#distance between defenders to line of pass
for index,value in df_filter.iterrows():
    sender_xy=value[['sender_x','sender_y']].values
    receiver_xy=value[['receiver_x','receiver_y']].values
    defender_xy=value[['defenders_xy']]
    
    for i in range(0,len(defender_xy)):
        dist_defender_lop.append(np.round(min(np.abs(np.cross(receiver_xy-sender_xy,defender_xy[i]-sender_xy)/LA.norm(receiver_xy-sender_xy))),2))
df_filter['dist_sr_closest_defender_lop']=dist_defender_lop
df_filter.head()
```

```python
from numpy import linalg as LA
dist_o_defender_lop=[]
dist_op_defender_lop=[]

count=0
#distance between defenders to line of pass for all possible passes
for index,value in df_filter.iterrows():
    sender_xy=value[['sender_x','sender_y']].values
    attacker_xy=value[['attackers_xy']].values.tolist()
    defender_xy=value[['defenders_xy']].values.tolist()
    dist_o_defender_lop=[]

    for i in range(0,len(attacker_xy[0])):
        dist_defender_lop=[]
        
        for j in range(0,len(defender_xy[0])):
            lop=np.abs(np.cross(attacker_xy[0][i]-sender_xy,defender_xy[0][j]-sender_xy)/LA.norm(attacker_xy[0][i]-sender_xy))
            dist_defender_lop.append(lop)
        
        dist_o_defender_lop.append(np.round((dist_defender_lop),2))
    dist_op_defender_lop.append(dist_o_defender_lop)

df_filter['dist_other_passes_defender_lop']=dist_op_defender_lop
df_filter.head()
```

```python
#Angle between sender and receiver
from math import atan2,degrees
angle_pass_players=[]

for index,value in df_filter.iterrows():
    attack_xy=value[['attackers_xy']].values.tolist()
    angle_pass_sr=[]

    for i in range(0,len(attack_xy[0])):
        angle=np.round(degrees(atan2((value['sender_y'] - attack_xy[0][i][1]), (value['sender_x'] - attack_xy[0][i][0])))%90,2)
        angle_pass_sr.append(angle)
    angle_pass_players.append(angle_pass_sr)

df_filter['angle_sr']=angle_pass_players
df_filter.head()
```
